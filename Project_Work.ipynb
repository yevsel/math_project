<h4 style="color:#333;font-weight:100">BRUTE FORCE ATTACK PREOCEDURE</h4>
<ul>
    <li>Pick an arbitrary or random password from the list of generated possible password combinations.</li>
    <li>Run 20 iterations.</li>
    <li>In each iteration calculate the time it takes to match that same password in the list generated.</li>
    <li>Then we return the average time taken for each password to be "hacked".</li>
</ul>

```python
def brute_force_attack(password_list):
    execution_times=[]
    a=[]
    for _ in range(20):
        random_password = choice(password_list)
        start_time = process_time()
        
        for password in password_list:
            if password==random_password:
                a.append([password,random_password])
                end_time = process_time()
                execution_times.append(end_time-start_time)
                break
    
    return (execution_times)
```

<h4 style="color:#333;font-weight:100;">Hacking (a)</h4>

```python
passwords = permutations(upper_cases,3)
execution_times = brute_force_attack(list(passwords))
collect_average_times.append(sum(execution_times)/len(execution_times))

print("Average Time Taken is",sum(execution_times)/len(execution_times),"seconds")
```
